```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from math import log
```

```python
def f(x, y):
    r=3**(-x**2 - y**2)
    return 1/(r+1)
```

```python
x1=np.linspace(-2,2,200)
```

```python
y1=np.linspace(-2,2,200)
```

```python
x1, y1=np.meshgrid(x1, y1)
```

```python
x1.shape
```

```python
y1.shape
```

```python
fig=plt.figure(figsize=(8, 8)).add_subplot(projection= '3d')
fig.set_xlabel('x', fontsize=20, c='red')
fig.set_ylabel('y', fontsize=20, c='blue')
fig.set_zlabel('z', fontsize=20, c='black')
fig.plot_surface(x1, y1, f(x1, y1), cmap='RdBu', alpha= 0.2)
plt.show()
```

```python
def fpx(x, y):
    
    
    
    r = 3**(-x**2 - y**2)
        
        
    return  2*x*log(3)*r / (r +1)**2 
```

```python
def fpy(x, y):
    
    
    r = 3**(-x**2 - y**2)
        
    return  2*y*log(3)*r / (r +1)**2 
```

```python
para=np.array([1.8, 1.1])
```

```python
para.shape
```

```python
para = np.array([1.8,1])

max_iter = 1000

multiplier = 0.1

values_array = para.reshape(1,2)


for n in range(max_iter):
    
    
    gradient_x = fpx(para[0], para[1])
    
    gradient_y = fpy(para[0], para[1])
    
    gradients = np.array([gradient_x, gradient_y])
    
    para = para - multiplier * gradients
    
    values_array = np.append(values_array, para.reshape(1,2), axis = 0 )
```

```python
fig=plt.figure(figsize=(8, 8)).add_subplot(projection= '3d')
fig.set_xlabel('x', fontsize=20, c='red')
fig.set_ylabel('y', fontsize=20, c='blue')
fig.set_zlabel('z', fontsize=20, c='black')
fig.plot_surface(x1, y1, f(x1, y1), cmap='RdBu', alpha= 0.2)
fig.scatter(values_array[:,0], values_array[:,1], f(values_array[:,0], values_array[:,1]), color = 'red', s = 10, alpha = 0.4)
plt.show()
```
