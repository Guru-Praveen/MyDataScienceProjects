# $ f(x) = x^2 + x + 1 $

```python
def f(x):
    return x ** 2 + x + 1
```

```python
x_1=np.linspace(-4,3,200)
x_1
```

```python
f(x_1)
```

```python
plt.figure(figsize =(7,6))
plt.plot(x_1,f(x_1), color = 'red', linewidth = 3)
plt.grid()
plt.show()
```

```python
def df(x):
    return 2 * x + 1
```

```python
plt.figure(figsize =(7,6))
plt.plot(x_1, df(x_1), color = 'navy', linewidth = 3)
plt.grid()
plt.show()
```

```python
plt.figure(figsize =(14,6))

plt.subplot(1, 2, 1)
plt.plot(x_1, f(x_1), color = 'red', linewidth = 3)
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x_1, df(x_1), color = 'navy', linewidth = 3)
plt.grid()

plt.show()
```

```python
new_x = 3
gamma = 0.1

for i in range(150):
    
    previous_x = new_x
    slope = df(previous_x)
    new_x = previous_x - gamma * slope
    
print(round(new_x,2))    
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-2,2,500)
```

```python
def g(x):
    return (x**4) - (4 * x**2) + 5 
```

```python
def g_der(x):
    return (4* x**3) - (8 * x)
```

```python
g(x)
```

```python
g_der(x)
```

```python
plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
plt.plot(x, g(x))

plt.subplot(1,2,2)
plt.plot(x,g_der(x))


plt.show()
```

```python
def grad_des(der_fun, ini, mul, no_iter, prec):
    new_x=ini
    lx=[new_x]
    slope=der_fun(new_x)
    sl_lst=[slope]
    for i in range(no_iter):
        prev=new_x
        grad=der_fun(prev)
        new_x=prev-mul * grad
        step_size = abs(new_x - prev)
        lx.append(new_x)
        sl_lst.append(der_fun(new_x))
        if step_size < prec:
            break
    return new_x, lx, sl_lst
```

```python
nx, l_x, sl = grad_des(g_der, 2, 0.01, 150, 0.001)
```

```python
print(nx)
```

```python
plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
plt.scatter(l_x, g(np.array(l_x)), s=80, c='red', alpha = 0.2)
plt.plot(x, g(x))
plt.xlabel()
plt.grid()

plt.subplot(1,2,2)
plt.scatter(l_x, sl, s=80, c='blue', alpha= 0.2)
plt.plot(x,g_der(x))
plt.grid()


plt.show()
```
