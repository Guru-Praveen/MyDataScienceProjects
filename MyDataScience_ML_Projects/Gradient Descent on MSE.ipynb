```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
x = np.array([[0.1, 1.2, 2.4, 3.2, 4.1, 5.7, 6.5]]).transpose()
y = np.array([[1.7, 2.4, 3.5, 3.0, 6.1, 9.4, 8.2]]).transpose()
```

```python
reg=LinearRegression()
reg.fit(x,y)
```

```python
plt.scatter(x, y)
plt.plot(x, reg.predict(x), c='red')
plt.show()
```

```python
reg.predict(x)
```

```python
mean_squared_error(y, reg.predict(x))
```

```python
def mse (y, y_hat):
    
    mse_calc = 1/7 * sum((y-y_hat)**2)
    
    return mse_calc
```

```python
mse(y, reg.predict(x))[0]
```

```python
y_hat = reg.intercept_[0] + reg.coef_[0][0]*x
y_hat 
```

```python
nr_thetas = 200
th_0 = np.linspace(start=-5, stop=5, num=nr_thetas)
th_1 = np.linspace(start=-5, stop=5, num=nr_thetas)
```

```python
th_0, th_1 = np.meshgrid(th_0, th_1)
```

```python
MSE=np.zeros((nr_thetas, nr_thetas))
```

```python
for i in range(nr_thetas):
    
    for j in range(nr_thetas):
        
        y_hat = th_0[i][j] + th_1[i][j]*x
        
        MSE[i][j] = mse(y, y_hat)
```

```python
fig=plt.figure(figsize=(16, 12)).add_subplot( projection='3d')

fig.set_xlabel('Theta_0', fontsize=20)

fig.set_ylabel('Theta_1', fontsize=20)

fig.set_zlabel('MSE - The Cost', fontsize=20)

fig.plot_surface(th_0, th_1, MSE, alpha = 0.4, cmap = cm.hot)

fig.azim = -45


plt.show()
```
