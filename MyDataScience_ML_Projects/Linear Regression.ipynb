```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('cost_revenue_clean.csv')
df
```

```python
x=pd.DataFrame(df,columns=['production_budget_usd'])
y=pd.DataFrame(df,columns=['worldwide_gross_usd'])
```

```python
df.describe()
```

```python
reg=LinearRegression()
reg
```

```python
f=reg.fit(x,y)
f
```

```python
res=reg.predict(x)
res
```

```python
plt.figure(figsize=(12,6))
plt.scatter(x,y,color='navy',s=80, alpha=0.3)
plt.plot(x,res,linewidth=4,color='red')
plt.show()
```
