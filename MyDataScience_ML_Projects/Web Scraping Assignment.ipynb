```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url='https://www.flipkart.com/search?sid=tyy%2C4io&otracker=CLP_Filters&p%5B%5D=facets.brand%255B%255D%3DAPPLE'
```

```python
text= requests.get(url)
```

```python
soup = bs(text.content, 'html.parser')
```

```python
data= soup.findAll('a', attrs = {'class': '_1fQZEK'})
```

```python
mob_name=[]
mob_price=[]
for i in data:
    name=i.find('div', attrs={'class': '_4rR01T'}).text
    price=i.find('div', attrs={'class': '_30jeq3 _1_WHN1'}).text
    mob_name.append(name)
    mob_price.append(price)
```

```python
df=pd.DataFrame({'Mobile Name': mob_name, 'Mobile Price': mob_price})
df
```
